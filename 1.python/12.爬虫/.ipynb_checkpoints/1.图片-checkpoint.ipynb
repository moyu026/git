#### 1.request发送请求，从服务器获取到数据
#### 2. BeautifulSoup来解析整个页面的源代码

```python
import requests
from bs4 import BeautifulSoup
```

```python
# 发送请求到服务器
resp = requests.get('http://www.umeituku.com/katongdongman/')
resp.encoding = "utf-8"
print(resp.text)
```

![jupyter](./file/1.png)

```python
# 解析html
main_page = BeautifulSoup(resp.text, "html.parser")
# 从页面中找到某些东西
# find() 找一个； find_all() 找所有
type_list = main_page.find("div", attrs={"class": "TypeList"})# 与上图框中的参数对应, 找到页面中所有图片对应的信息
print(type_list)
```

![jupyter](./file/2.png)

```python
a_list = main_page.find("div", attrs={"class": "TypeList"}).find_all("a", attrs={"class": "TypeBigPics"})# 取带图片链接的参数
print(a_list)
```

```python
n = 0
for a in a_list:
#     print(a.get("href"))
    # 发送请求到子页面，进入到子页面中
    href = a.get("href")
    resp1 = requests.get(href)
    resp1.encoding = "utf-8"
    child_page = BeautifulSoup(resp1.text,"html.parser")
    src = child_page.find("div", attrs={"class": "ImageBody"}).find("img").get('src')
    print(src)
    
    # 发送请求到服务器，把图片保存到本地
    # 创建文件
    f = open("./file/tu_%s.jpg"%n, "wb")
    f.write(requests.get(src).content) # 拿出图片的数据，不是文本信息
    n += 1
```

![jupyter](./file/3.png)

```python
# 获取套图的下一页
for a in a_list:
#     print(a.get("href"))
    # 发送请求到子页面，进入到子页面中
    href = a.get("href")
    resp1 = requests.get(href)
    resp1.encoding = "utf-8"
    child_page = BeautifulSoup(resp1.text,"html.parser")
    text = child_page.find("div", attrs={"class": "ImageBody"}).find_next('script').text
    print(text)
# Next("1","1","204096","st","204096","") 第一个1是当前页数，第二个1是总页数
```
